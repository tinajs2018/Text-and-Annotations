```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df=pd.read_csv('glass (1).csv')
df.head()
```

```python
df['Mg'].describe()
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(15,5))
plt.plot(df['Mg'])
# plt.text(102,2.115,'the lowest mg')
plt.text(102, 2.115, "lowest mg",color="blue")
plt.text(0,4.49,"the highest",color='blue')

plt.title('The components  of glass')
plt.show()
```

```python
df['Ca'].describe()
```

```python
fig, ax = plt.subplots(figsize=(15, 4))
df['Ca'].plot(ax=ax,label='ca')
ax.annotate("Lowest mg used", xy=(106, 16.190), xycoords='data',
xytext=(50, -30), textcoords='offset points',
arrowprops=dict(arrowstyle="->",
connectionstyle="arc3,rad=-0.2"))
plt.title('THe levels of calcium')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['Ca'],label='Ca')
plt.annotate("Lowest mg used", xy=(106, 16.190), xycoords='data',
xytext=(50, -30), textcoords='offset points',
arrowprops=dict(arrowstyle="->",
connectionstyle="arc3,rad=-0.2"))
plt.show()
```
